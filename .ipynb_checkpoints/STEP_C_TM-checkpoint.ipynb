### Template Matching

```python
import numpy as np
import imutils
import cv2
from matplotlib import pyplot as plt
import math
```

Template Matching is a method for searching and finding the location of a model (or template) image in a scene (or larger) image. It simply slides the model image over the scene image (as in 2D convolution) and compares the model and patch of scene image under the model image.

```python
def templateMatch(scene, model):
    #scale the model image to make it comparable with the scene image.
    best_match = None
    rectangles = []
    
    for scale in np.linspace(0.05, 0.5, 40):

        # Temporary resize the model image using scale level
        test_resized_model = imutils.resize(model, width = int(model.shape[1] * scale))
    
        # call matchTemplate function using cv2.TM_CCOEFF_NORMED method which gives maximum value as the best match
        res_t = cv2.matchTemplate(scene, test_resized_model, cv2.TM_CCOEFF_NORMED)
        
        min_val_t, max_val_t, min_loc_t, max_loc_t = cv2.minMaxLoc(res_t)
        
        if best_match is None or max_val_t >= best_match:
            #NOT USED IDEAL_SCALE
            #ideal_scale = 0 #scale
            h = test_resized_model.shape[0]
            w = test_resized_model.shape[1]
            best_match = max_val_t #max_loc_t
     
    #get rid of incorrect detections
    if w >= 45 and best_match >= 0.4:
        #resize and match template 
        resized_model = cv2.resize(model, (w, h), interpolation = cv2.INTER_AREA)
        res = cv2.matchTemplate(scene, resized_model, cv2.TM_CCOEFF_NORMED)
        # Maximum value gives the best match in cv2.TM_CCOEFF_NORMED method.
        # We can get mininum and and maximum values and minimum and maximum locations from minMaxLoc() function.
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
        # Need to detect other approximate matches. So we must find the threshold values to detect other matches.
        if max_val <= 0.5:
            threshold = 0.4
        elif max_val < 0.6:
            threshold = 0.5
        elif max_val < 0.7:
            threshold = 0.6
        elif max_val < 0.8:
            threshold = 0.7
        else:
            threshold = 0.7
            
        # Detect x,y points
        yloc, xloc = np.where(res >= threshold)
    
        # Most of the time, it detect duplicate rectangles for one product in the scene image
        # So we need to group the duplicate rectangles.
        for (x, y) in zip (xloc, yloc):
            rectangles.append([int(x), int(y), int(w), int(h)])
            # Store two times of each value for group the rectangles
            rectangles.append([int(x), int(y), int(w), int(h)])
        
        rectangles, widths = cv2.groupRectangles(rectangles, 1, 0.2)
        
        return rectangles
```

```python
for scene in range(5):

    s_img = 'scenes/h' + str(scene + 1) + '.jpg'

    scene_img = cv2.imread(s_img, cv2.IMREAD_UNCHANGED)
    
    scene_rgb = cv2.cvtColor(scene_img, cv2.COLOR_BGR2RGB)
    
    #uncomment to display rgb images
    #plt.imshow(scene_rgb)
    #plt.show()
    
    for m in range(27):
        
        m_img = 'models/' + str(m) + '.jpg'
        model = cv2.imread(m_img, cv2.IMREAD_UNCHANGED)
        
        # Some model images have very large resolutions -> reduce the image size to 50%.
        if model.shape[0] >= 1000:
            model = cv2.resize(model, (int(model.shape[1] * 0.5), int(model.shape[0] * 0.5)), interpolation = cv2.INTER_AREA)
        
        rectangles = templateMatch(scene_img, model)
        
        if rectangles is not None:
            print('Product {} - {} instance/s found'.format(m, len(rectangles)))

            for (x, y, w, h) in rectangles:
                print("\t", "Position ({},{}), width: {} px, height: {} px".format(x,y, w, h))
                cv2.rectangle(scene_rgb, (x, y), (x + w, y + h), (0,255,0), 2)
        else:
            print('Product ',m,' - no instance found')
                
    plt.figure(figsize=(20, 7))
    plt.imshow(scene_rgb)
    plt.show()
```
